```python
import requests
from bs4 import BeautifulSoup

# Function to extract title from HTML content
def extract_title_from_html(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    title_tag = soup.title
    if title_tag:
        return title_tag.text.strip()
    else:
        return None

# Function to download PDF content
def download_pdf(url, filename):
    with open(filename, 'wb') as pdf_file:
        response = requests.get(url)
        pdf_file.write(response.content)

# Function to handle URLs
def handle_url(url):
    response = requests.head(url)
    content_type = response.headers.get('Content-Type', '')

    # Ignore character encoding information in Content-Type header
    content_type = content_type.split(';')[0]

    if content_type == 'text/html':
        # If the content type is HTML
        html_content = requests.get(url).content
        title = extract_title_from_html(html_content)
        if title:
            print("Title:", title)
        else:
            print("No title found in the HTML.")
    
    elif content_type == 'application/pdf':
        # If the content type is PDF
        pdf_filename = 'downloaded_pdf.pdf'  # You can customize the filename here
        download_pdf(url, pdf_filename)
        print("PDF downloaded as", pdf_filename)
    
    else:
        # Handle other content types if needed
        print("Unsupported content type:", content_type)

# Example usage
if __name__ == "__main__":
    url = "https://example.com"  # Replace with the URL of the website you want to handle
    handle_url(url)
```
